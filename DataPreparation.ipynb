```python
import pickle
with open('ctypes_data_all_missions.pkl', 'rb') as f:
    df = pickle.load(f)
```

```python
# Droping indexes.
df.reset_index(drop=True, inplace=True)
df
```

```python
import re
import pandas as pd

# Find first entrance word to string and convert it to lower case.
types_string = df['CTypes'].str.split(",")
types_string = pd.Series([tuple({re.findall("\w+", w)[0].lower() for w in l if len(re.findall("\w+", w)) > 0}) for l in types_string])
types_string.reset_index(drop=True, inplace=True)
# Display unique sets of cloud types.
set(types_string)
```

```python
print(len(types_string))
```

```python
# Base 10 types:
ctypes = ['ci', 'cc', 'cs', 'ac', 'as', 'sc', 'st', 'ns', 'cu', 'cb']
```

```python
import numpy as np
base_df = df.copy()
base_df[ctypes] = pd.DataFrame([[0] * 10], index=df.index)
```

```python
for ct in ctypes:
    for i, cd in enumerate(types_string):
        if ct in cd:
            base_df.at[i, ct] = 1
```

```python
base_df['sum'] = base_df[ctypes].sum(axis=1)
# Add class "not recognized"
base_df['nr'] = base_df['sum'].apply(lambda x: 1 if x == 0 else 0)
# exclude the data with class "not recognized"
base_df = base_df[base_df['nr'] != 1]
base_df
```

```python
# Statistics of classes.
base_df.sum()
# The most frequent is "sc" and the rearest is "cb"
```

```python
import os

# Find all pics of clouds from expeditions.
storage_path = "/storage/kubrick/krinitsky/DASIO"

# Waiting 5 mins...
files_path = []
for addr, dirs, files in os.walk(storage_path):
    if not dirs:
        for file in files:
            if file.endswith(".jpg"):
                files_path.append(os.path.join(addr, file))
```

```python
base_df.loc[:, 'dt'] = pd.to_datetime(base_df.loc[:, 'dt'])
```

```python
for ct in ctypes:
    base_df.loc[:, ct] = base_df.loc[:, ct].astype("category")
```

```python
base_df.groupby(['sum']).count().CTypes
# The number of classes in the snapshot is distributed in a natural order.
```

```python
files_path = set(files_path)
```

```python
pics = list()
for path in files_path:
    dict_p = dict()
    dict_p['Mission'] = os.path.basename(os.path.dirname(os.path.dirname(path)))
    dict_p['Date_time'] = pd.to_datetime(os.path.basename(path)[4:-10], format='%Y-%m-%dT%H-%M-%S')
    dict_p['Full_path'] = path
    pics.append(dict_p)
```

```python
df_pics = pd.DataFrame(pics)
```

```python
df_pics = df_pics.drop_duplicates()
df_pics["Mission"].replace('snapshots-2015-12-28', 'ANS31', inplace=True)
df_pics.shape
```

```python
df_pics.sort_values(by=['Date_time'], inplace=True)
base_df.sort_values(by=['dt'], inplace=True)
```

```python
full_df = pd.merge_asof(df_pics, base_df, left_on='Date_time', right_on='dt', tolerance=pd.Timedelta(minutes=5))
full_df.dropna(subset=['dt'], inplace=True)
```
